```python
Finding outliers
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_limit) | (df['price'] > upper_limit)]

print("Total Outliers =", len(outliers))
```
